```python
# Using IF statements with user-defined functions
```

```python
#Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Creating  a new column, price_range and, applying a string label for every row based on its price
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
# IF statements with loc() function
```

```python
# Creating a new column, price_range_loc and assigning the string label by locating the prices using loc() function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df.head()
```

```python
# Creating a new column, price_range_loc on df_orders_products_merged dataframe to check efficiency of the loc() function on the full dataframe 

df_orders_products_merged.loc[df_orders_products_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
df_orders_products_merged.loc[(df_orders_products_merged['prices'] <= 15) & (df_orders_products_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
df_orders_products_merged.loc[df_orders_products_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Find how busy each day of the week is
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Day 0, Saturday is the busiest and Day 4, Wednesday is the least busy day.
```

```python
# loop through every row and assign the string label according to the value of the "orders_day_of_week" column.
# store it in a list

result = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create a new column "busiest_day" and assign the list created from the loop with the values for the new column

df_orders_products_merged['busiest_day'] = result
```

```python
df_orders_products_merged['busiest_day'].value_counts(dropna  = False)
```

```python
# STEP 2. To find Busiest days(2 busiest days and 2 slowest days)
```

```python
# loop through every row and assign the string label according to the value of the "orders_day_of_week" column.
# store it in a list

result_busiest_days = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if (value == 0) or (value == 1):
    result_busiest_days.append("Busiest days")
  elif (value == 4) or (value == 3):
    result_busiest_days.append("Slowest days")
  else:
    result_busiest_days.append("Regularly busy")
```

```python
result_busiest_days
```

```python
# create a new column "busiest_days" and assign the list created from the loop with the values for the new column

df_orders_products_merged['busiest_days'] = result_busiest_days
```

```python
df_orders_products_merged['busiest_days'].value_counts(dropna  = False)
```

```python
#STEP 4. Find busiest hours of the day
```

```python
# Find how busy each hour of the day is
df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Most orders  - periods of time where more than 2 million orders were made
# Average orders - periods of time between 1 million and 2 million orders
# Fewest orders  - periods of time where less than 1 million orders were made
```

```python
# loop through every row and assign the string label according to the value of the "order_hour_of_day" column.
# store it in a list

result_busiest_period_of_day = []

for value in df_orders_products_merged["order_hour_of_day"]:
  if (value in [10,11,14,15,13,12,16,9,17]):
    result_busiest_period_of_day.append("Most orders")
  elif (value in [3,4,2,5]) :
    result_busiest_period_of_day.append("Fewest orders")
  else:
    result_busiest_period_of_day.append("Average orders")
```

```python
result_busiest_period_of_day
```

```python
# create a new column "busiest_period_of_day" and assign the list created from the loop with the values for the new column

df_orders_products_merged['busiest_period_of_day'] = result_busiest_period_of_day
```

```python
#STEP 5. Print the frequency for this column
df_orders_products_merged['busiest_period_of_day'].value_counts(dropna  = False)
```
