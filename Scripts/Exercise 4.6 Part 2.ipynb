# Combinind Data and Exporting - Part 2

```python
#Importing libraries
import numpy as np
import pandas as pd
import os
```

```python
# Setting the path for the datasets
path = r'C:\\Users\\cpaul\\Documents\\Career Foundry\\Instacart Basket Analysis'

#STEP 3: Import dataframe

# Importing datasets, orders_products_combined and products_checked
df_orders_products_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
df_products = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

```python
# STEP 4: Check shape of the imported dataframe

df_orders_products_combined.shape
```

```python
df_orders_products_combined.head()
```

```python
# Drop existing _merge column from df_orders_products_combined
df_orders_products_combined = df_orders_products_combined.drop(columns = ['_merge'])
```

```python
df_orders_products_combined.head()
```

```python
# Checking df_products dataframe
df_products.shape
```

```python
df_products.head()
```

```python
# STEP 5: Determine a suitable way to combine the orders_products_combined dataframe with df_products

# Merge these two dataframes using product_id as the key, since it matches between the two dataframes
df_merged = df_orders_products_combined.merge(df_products, on = 'product_id', indicator = True)
```

```python
df_merged.shape
```

```python
df_merged.head()
```

```python
# Drop Unnamed columns from df_merged dataframe
df_merged = df_merged.drop(columns = ['Unnamed: 0.1','Unnamed: 0_x','Unnamed: 0_y'])
```

```python
df_merged.head()
```

```python
# STEP 6: Confirm the results of the merge flag
df_merged['_merge'].value_counts()
```

```python
# The datasets are a full match.
```

```python
#STEP 7: Export the newly created dataframe as orders_products_merged 

#Export it in pkl format
df_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
