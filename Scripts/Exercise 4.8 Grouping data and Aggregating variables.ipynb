# STEP 4: Creating loyalty flags for existing customers

```python
# Create the max_order column which stores the the maximum order number for each user. 
# So, the dataframe is first grouped by user_id and then the transform() function is applied on the order_number column with np.max argument

df_orders_products_merged_updated['max_order'] = df_orders_products_merged_updated.groupby(['user_id'])['order_number'].transform(np.max)

df_orders_products_merged_updated.head(20)
```

```python
# Create a loyalty flag using the loc() function

df_orders_products_merged_updated.loc[df_orders_products_merged_updated['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_orders_products_merged_updated.loc[(df_orders_products_merged_updated['max_order'] <= 40) & (df_orders_products_merged_updated['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_orders_products_merged_updated.loc[(df_orders_products_merged_updated['max_order'] <= 10), 'loyalty_flag'] = 'New customer'

# Check the frequency of the loyalty_flag column

df_orders_products_merged_updated['loyalty_flag'].value_counts(dropna = False)
```

```python
df_orders_products_merged_updated[['user_id','loyalty_flag','order_number']].head(60)
```

# STEP 5: Determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Check the basic statistics of the product prices for each loyalty category
df_orders_products_merged_updated.groupby('loyalty_flag').agg({'prices': ['mean','min','max','median']})
```

# The mean of the product price for Loyal customers seem to be lower than the regular customers, also there are more regular customers than loyal customers. The number of New customers are lesser than the other two groups, so we cannot fully take into account the mean of product price of this group even though the mean is the highest.

# STEP 6: Target different types of spenders in their marketing campaigns

```python
# Create a spending flag for each user based on the average price across all their orders

df_orders_products_merged_updated['mean_price'] = df_orders_products_merged_updated.groupby(['user_id'])['prices'].transform(np.mean)

df_orders_products_merged_updated.head(20)
```

```python
# Create a spender_type flag using the loc() function

df_orders_products_merged_updated.loc[df_orders_products_merged_updated['mean_price'] >= 10, 'spender_type_flag'] = 'High spender'
df_orders_products_merged_updated.loc[(df_orders_products_merged_updated['mean_price'] < 10), 'spender_type_flag'] = 'Low spender'

# Check the frequency of the loyalty_flag column

df_orders_products_merged_updated['spender_type_flag'].value_counts(dropna = False)
```

# STEP 7: Determine frequent versus non-frequent customers  to send relevant notifications within the app

```python
# Create a frequency flag for each user based on the median of days_since_previous_order across all their orders

df_orders_products_merged_updated['median_days_since_previous_order'] = df_orders_products_merged_updated.groupby(['user_id'])['days_since_previous_order'].transform(np.median)

df_orders_products_merged_updated.head(20)
```

```python
# Create a frequency flag using the loc() function

df_orders_products_merged_updated.loc[df_orders_products_merged_updated['median_days_since_previous_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
df_orders_products_merged_updated.loc[(df_orders_products_merged_updated['median_days_since_previous_order'] <= 20) & (df_orders_products_merged_updated['median_days_since_previous_order'] > 10), 'frequency_flag'] = 'Regular customer'
df_orders_products_merged_updated.loc[(df_orders_products_merged_updated['median_days_since_previous_order'] <= 10), 'frequency_flag'] = 'Frequent customer'

# Check the frequency of the frequency_flag column

df_orders_products_merged_updated['frequency_flag'].value_counts(dropna = False)
```

```python
# Checking the Nan values
print(df_orders_products_merged_updated[df_orders_products_merged_updated['frequency_flag'].isnull()])
```

# This user with user_id 159838 has only one order and days_since_previous_order column is null due to this, and, the median cannot be calculated. So, I would like to flag this customer as a Non-frequent customer.

```python
# Updating the frequncy_flag for this customer
df_orders_products_merged_updated['frequency_flag'].fillna("Non-frequent customer", inplace = True)

df_orders_products_merged_updated['frequency_flag'].value_counts(dropna = False)
```
